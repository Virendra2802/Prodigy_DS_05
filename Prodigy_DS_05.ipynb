Task-5: This project analyzes road traffic accident data to identify patterns related to road conditions, weather, driver demographics, and time of day. The dataset used focuses on accident records from 2017 to 2020 and includes details about drivers, casualties, vehicles, and accident severity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df.columns
df['Weather_ocnditions'] = df['Weather_conditions'].fillna('Unknown')
df['Road_surface_type'] = df['Road_surface_type'].fillna('Unknown')
```

```python
print(df[['Weather_conditions', 'Road_surface_type']].head())
```

```python
top_weather = df['Weather_conditions'].value_counts().head(10)

plt.figure(figsize=(9,5))
sns.barplot(x=top_weather.index, y=top_weather.values, palette='viridis')
plt.title('Top Weather Conditions during Accidents')
plt.ylabel('Number of Accidents')
plt.ylim(0, 10000)
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Road_surface_type', data=df, palette='flare', order=df['Road_surface_type'].value_counts().index)
plt.title("Accidents by Road Surface Type")
plt.ylabel("Road Surface Type")
plt.xlabel("Frequency")
plt.show()
```

```python

plt.figure(figsize=(10,6))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='Day_of_week', data=df, order=order, palette='viridis')
plt.title('Number of Accidents by Day of Week')
plt.ylabel('Accident Count')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Casualty_severity', data=df, palette='rocket')
plt.title('Casualty Severity Distribution')
plt.xlabel('Casualty Severity')
plt.ylabel('Number of Cases')
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_causes = df['Cause_of_accident'].value_counts().head(10)
sns.barplot(y=top_causes.index, x=top_causes.values, palette='magma')
plt.title('Top 10 Causes of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Cause of Accident')
plt.show()
```

```python
plt.figure(figsize=(12,6))
vehicle_counts = df['Type_of_vehicle'].value_counts().head(10)
sns.barplot(x=vehicle_counts.values, y=vehicle_counts.index, palette='cool')
plt.title('Top 10 Vehicle Types Involved in Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Vehicle Type')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Driving_experience', hue='Accident_severity', data=df, palette='cubehelix')
plt.title('Driving Experience vs Accident Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
